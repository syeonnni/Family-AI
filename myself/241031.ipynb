```python
mega = "https://www.mega-mgccoffee.com/store/find/store.php?"

payload = {'sigungu': '강남구'}

bs = BeautifulSoup(requests.get("https://www.mega-mgccoffee.com/store/find/store_area_search.php?store_area_name=%EC%84%9C%EC%9A%B8").text)
mega_list = []
for code in [x.text.strip() for x in bs.findAll("li", class_="store_area_search_list")]:
    payload['sigungu'] = code
    mega_list.extend(requests.post(mega, data=payload).json()['positions'])
```
