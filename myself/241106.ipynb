```python
import fnmatch
import shutil
male=[]
female=[]
for roots, dirs, files in os.walk("../encore/"):
    for file in files:
        print(f"{roots}{file}")
        if file.split(".")[-1] == "txt":
            if file.split("_")[-1][0] in ['1', '3']:
                with open(f"{roots}{file}", "r", encoding='utf-8') as f:
                    male.append(f.read())
                shutil.move(f"{roots}{file}", f"./tmp/male/{file}")
            elif file.split("_")[-1][0] in ['2', '4']:
                with open(f"{roots}{file}", "r", encoding='utf-8') as f:
                    female.append(f.read())
                shutil.move(f"{roots}{file}", f"./tmp/female/{file}")
```

```python
import numpy as np
a = list(map(lambda x : int(x.replace(",", "")), female) )
```

```python
#테이블 삭제 
import pymysql
con = pymysql.connect(host='127.0.0.1', user='encore', passwd='1234', port=3306,
                database='encore')

cursor = con.cursor()

sql = "drop table people";

con.commit()
cursor.execute("show tables")
```

```python
#내가 짠 코드 (여성,남성 데이터를 DB테이블에 저장)
insert_sql = """INSERT INTO people VALUES (%s, %s, %s)"""

for roots, dirs, files in os.walk("./tmp/female/"):
    for file in files:
        if file.find("txt") > -1:
            file = file.split(".")[0]
            cursor.execute(insert_sql, [file.split(".")[0].split("_")[0], file.split(".")[0].split("_")[1]+'-'+file.split(".")[0].split("_")[2], 
                                        f"{roots}".split("/")[2]])

for roots, dirs, files in os.walk("./tmp/male/"):
    for file in files:
        if file.find("txt") > -1:
            file = file.split(".")[0]
            cursor.execute(insert_sql, [file.split(".")[0].split("_")[0], file.split(".")[0].split("_")[1]+'-'+file.split(".")[0].split("_")[2], 
                                        f"{roots}".split("/")[2]])


con.commit()
```
