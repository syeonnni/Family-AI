```python
total = []

for file in glob.glob("./TL1/*.json"):
    f = open(file, 'r')
    tmp = json.loads(f.read())
    article = tmp['Meta(Refine)']['passage']
    summary = tmp['Annotation']['summary1']
    total.append({'article' : article, 'summary' : summary})
```

```python
pd.DataFrame(total).to_csv("./")
```
