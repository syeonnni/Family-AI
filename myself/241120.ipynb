```python
import re
total = []
p = re.compile("[0-9]+")

for file in os.listdir("./seoul"):
    if (file[-3:] == "csv"):
        #print(file)
        try:
            tmp = pd.read_csv(f"./seoul/{file}", encoding='utf-8', low_memory=False)
        except Exception as e:
            tmp = pd.read_csv(f"./seoul/{file}", encoding='cp949', low_memory=False)
        tmp['날짜'] = p.findall(file)[0]
        total.append(tmp)
```

```python
seoul_df = pd.concat(total, ignore_index=True)
```

```python
sorted(seoul_df.날짜.unique())
```
