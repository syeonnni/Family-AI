```python
import glob
from konlpy.tag import Komoran, Okt

komoran = Komoran(userdic='./user_dic.txt')
```

```python
komoran.pos("오징어 게임 재미 있습니까? 괜찮아 괜찮아 괜찮아 영화는?")
```

```python
total = []
for file in glob.glob("./2024-12-09/*.txt"):
    with open(glob.glob("./2024-12-09/*.txt")[0], "r") as f:
        for x in f:
            for y in komoran.pos(x):
                # print(x)
                if y[1] in ['NNP', 'VV'] and len(y[0]) > 1:
                    print(y[0])
```

```python
komoran.pos("오징어 게임 개봉")
```

```python
total = []
for file in glob.glob("./2024-12-09/*.txt"):
    with open(file, "r") as f:
        try:
            total.extend([y[0] for x in f for y in komoran.pos(x) if y[1] in ['NNP', 'VV'] and len(y[0]) > 1])
        except:
            pass
```

```python
from collections import Counter
```

```python
Counter(total)
```

```python
"오징어 게임" in Counter(total).keys()
```

```python
import pickle
with open("./cnt_word.pkl", "wb") as f:
    pickle.dump(Counter(total),f)
```

```python
import pickle
with open("./cnt_word.pkl", "rb") as f:
    total_cnt = pickle.load(f)
```
