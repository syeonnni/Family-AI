```python
krx = "http://data.krx.co.kr/comm/bldAttendant/getJsonData.cmd"

payload = {"bld": "dbms/MDC/STAT/standard/MDCSTAT01901",
"locale": "ko_KR",
"mktId": "ALL",
"share": "1",
"csvxls_isNo": "false"}

head = {"user-agent":
"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/130.0.0.0 Safari/537.36",
"host" : "data.krx.co.kr",
"origin": "http://data.krx.co.kr",
"referer" :"http://data.krx.co.kr/contents/MDC/MDI/mdiLoader/index.cmd?menuId=MDC0201020201"
       }

r = requests.post(krx, data=payload, headers=head)

r.json()
```

```python
r.json()['OutBlock_1']
```

```python
master = [x['ISU_SRT_CD'] for x in r.json()['OutBlock_1'] if x['MKT_TP_NM'] in ["KOSDAQ", "KOSPI"]]
```

```python
master
```

```python
import pickle
with open("./krx.pkl", "wb") as f: 
    pickle.dump(master,f)
```
