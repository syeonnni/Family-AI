## OPEN AI가 제공해주는 임베딩 모듈 활용하기(유료결제)

```python
#openai 인증키를 cmd -> keys.txt -> 입력 후 저장 / key 값 출력
with open("./openai/keys.txt", "r") as f:
    keys = f.read().strip() 
```

```python
#openai 연동
import openai 
client = openai.OpenAI(api_key = keys)
```

```python
def get_embedding(text, engine):
    #openai 서버에서 res로 데이터 보내줌 
    res = client.embeddings.create(
        input = text
        model = engine
    )
    #보내준 데이터 받기 
    return [d.embedding for d in list(res.data)]
```

```python
a = get_embeddings("안녕하세요. 오늘은 월요일입니다.", "text-embedding-ada-002")
```

```python
b = get_embeddings("안녕하세요. 오늘은 금요일입니다.", "text-embedding-ada-002")
```

```python
#유사도 구하기
import numpy as np 
def cosine_similarity(vec1, vec2):
    #내적
    dot_result = np.dot(vec1, vec2)
    #벡터의 크기로 나눠주기 -> 크기는 numpy norm을 활용 
    dot_result / (np.linalg.norm(vec1) * np.linalg.norm(vec2))
```

```python
cosine_similarity(np.array(a[0]), np.array(b[0]))
```

```python
import chromadb, json
with open("./openai/ko_nia_normal_squad_all.json", "r") as f:
    json_text = f.read() 
```

```python
json_dataset = json.loads(json_text)
```

```python
json_dataset['data'][0]['paragraphs'][0]['qas'][0]['question']
```

```python
json_dataset['data'][0]['paragraphs'][0]['qas'][0]['answers'][0]['text']
```

```python
dataset_list = []
for idx, data in enumerate(json_dataset['data']):
    for qa in data['paragraphs'][0]['qas']:
        dataset = {}
        dataset['question'] = qa['question']
        dataset['answer']   = qa['answers'][0]['text']
        dataset_list.append(dataset)
```

```python
import pandas as pd
df = pd.DataFrame(dataset_list)
```

```python
df
```

```python
client = chromadb.PersistentClient()

answers = client.create_collection(
    name = "playdata"
)
```

```python
df = df.iloc[:10000]
```

```python
df.shape
```

```python
from tqdm import tqdm 
ids = [] 
metadatas = []
embeddings = []
for idx, row in tqdm(enumerate(df.iterrows())):
    query = row[1].question
    answer = row[1].answer
    
    metadata = {
        'query' : query,
        'answer' : answer
    }
    embedding = model.encode(query, normalize_embeddings=True)
    ids.append(str(idx))
    metadatas.append(metadata)
    embeddings.append(embedding)
```

```python
chunk_size = 1024 
total_chunks = len(embeddings) // chunk_size + 1
```

```python
for chunk_idx in tqdm(range(total_chunks)):
    start_idx = chunk_idx * chunk_size
    end_idx = (chunk_idx +1) * chunk_size

    chunk_embeddings = embeddings[start_idx:end_idx]
    chunk_ids = ids[start_idx:end_idx]
    chunk_metadatas = metadatas[start_idx:end_idx]

    answers.add(embeddings=chunk_embeddings, ids=chunk_ids, metadatas=chunk_metadatas)
```

```python
answers.query(
    query_embeddings=model.encode("이번 포럼 주제는 뭐야?", normalize_embeddings = True).tolist(),
    n_results=3
)
```

```python
import chromadb
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('jhgan/ko-sroberta-multitask')

client = chromadb.PersistentClient()

answer = client.get_collection("playdata")

while(True):
    text = input("입력하세요 : ")
    if text == 'Q':
        break
    result = answer.query(
            query_embeddings=model.encode(text, normalize_embeddings=True).tolist(),
            n_results=3)

    for y in [x['answer'] for x in result['metadatas'][0]]:
        print(y)
```
