```python
kbo = kbo[~kbo['신장/체중'].apply(lambda x : len(x) < 9)] 
```

```python
kbo['신장'] = kbo['신장/체중'].apply(lambda x : int(x.split("/")[0].replace("cm", "")))
```

```python
kbo['체중'] = kbo['신장/체중'].apply(lambda x : int(x.split("/")[-1].replace("kg", "")))
```

```python
import re
p = re.compile("[0-9]+") 
```

```python
kbo['신장/체중'].apply(lambda x : p.findall(x.split("/")[0]))
```

```python
kbo['신장/체중'].apply(lambda x : p.findall(x))
```

```python
#int형으로 변환
kbo['신장'] = kbo['신장/체중'].apply(lambda x : list(map(int, p.findall(x)))[0])
```

```python
kbo['체중'] = kbo['신장/체중'].apply(lambda x : list(map(int, p.findall(x)))[1])
```

```python
kbo.describe()
```

```python
#drop은 행과 열을 모두 포함 
kbo.drop("index", axis=1)
```

```python
#inplace 함수는 변경한 데이터를 바로 적용 
#index 없는 테이블에 drop 함수 적용하면 바로 에러
#axis 0은 행 / 1은 열 
kbo.drop("index", axis=1, inplace = True)
```
