```python
#default 매개변수는 뒤에서부터 채워야한다. type_의 디폴트값을 K로 설정 
def trans(x, type_="K"):
    """
    마일 사용 -> M
    킬로 사용 -> K 
    """
    type_ = type_.upper()
    if type_ == "M":
        return x / 1.60934
    elif type_ == "K":
        return x * 1.60934
    else:
        print("사용법숙지")
```

```python
trans(30, 'm')
```
