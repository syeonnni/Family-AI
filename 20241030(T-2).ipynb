```python
import requests 
import csv

def get_stock(code, start_date, end_date, type_='day'):
    url  = f"https://m.stock.naver.com/front-api/external/chart/domestic/info?symbol={code}&requestType=1&startTime={start_date}&endTime={end_date}&timeframe={type_}"
    r = requests.get(url)
    data = eval(r.text.strip())
    with open("./stock/{}.csv".format(code), "w", encoding='utf-8-sig') as f:
        w_stock = csv.writer(f)
        w_stock.writerows(data)
```
